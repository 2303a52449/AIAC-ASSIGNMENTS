````python
def calculate_student_grade(student_marks):
    """
    Calculates the total marks, average marks, and final grade for a student,
    with robust input validation.

    Args:
        student_marks (list): A list of numerical marks for the student.

    Returns:
        str: A string indicating the total marks, average marks, and final grade,
             or an error message if the input is invalid or the list is empty.
    """

    # Validate if the input is a list
    if not isinstance(student_marks, list):
        return "Error: Input must be a list of marks."

    # Check if the student_marks list is empty to prevent ZeroDivisionError
    if not student_marks:
        return "Error: Student marks list cannot be empty."

    # Validate each mark in the list
    for mark in student_marks:
        if not isinstance(mark, (int, float)) or not (0 <= mark <= 100):
            return f"Error: Each mark must be a number between 0 and 100. Invalid mark: {mark}"

    # Calculate the sum of all marks to get the total marks
    total_marks = sum(student_marks)
    # Calculate the average marks by dividing total marks by the number of marks
    average_marks = total_marks / len(student_marks)

    # Determine the grade based on the average marks using a series of conditional statements
    if average_marks >= 90:
        grade = 'A'
    elif average_marks >= 80:
        grade = 'B'
    elif average_marks >= 70:
        grade = 'C'
    elif average_marks >= 60:
        grade = 'D'
    else:
        grade = 'F'

    # Return a formatted string with the calculated total marks, average marks, and grade
    return f"Total Marks: {total_marks}, Average Marks: {average_marks:.2f}, Grade: {grade}"

# Example usage with valid inputs:
print("--- Valid Inputs ---")
print(calculate_student_grade([85, 92, 78, 65, 90]))
print(calculate_student_grade([50, 55, 45]))
print(calculate_student_grade([100, 100, 100]))

# Example usage with invalid inputs:
print("\n--- Invalid Inputs ---")
print(calculate_student_grade("not a list")) # Test with non-list input
print(calculate_student_grade(123))         # Test with non-list input
print(calculate_student_grade([]))          # Test with an empty list
print(calculate_student_grade([80, 'ninety', 70])) # Test with non-numeric mark
print(calculate_student_grade([80, 105, 70]))    # Test with mark > 100
print(calculate_student_grade([80, -5, 70]))     # Test with mark < 0
````

````python
def factorial(n):
    """
    Calculates the factorial of a non-negative integer.

    The factorial of a non-negative integer n, denoted by n!, is the product
    of all positive integers less than or equal to n.
    For example, 5! = 5 * 4 * 3 * 2 * 1 = 120.
    By definition, 0! is 1.

    Args:
        n (int): A non-negative integer for which to calculate the factorial.

    Returns:
        int: The factorial of n.

    Raises:
        ValueError: If n is a negative integer.
    """
    # Input validation: Ensure n is a non-negative integer
    if not isinstance(n, int) or n < 0:
        raise ValueError("Input must be a non-negative integer.")

    result = 1
    # Iterate from 1 up to n (inclusive) to calculate the product
    for i in range(1, n + 1):
        result *= i  # Multiply result by the current number
    return result

# Example usage:
print(f"Factorial of 5: {factorial(5)}")
print(f"Factorial of 0: {factorial(0)}")
# print(f"Factorial of -1: {factorial(-1)}") # This would raise a ValueError
````

    *   The `Returns` section describes what the function outputs (its type and meaning).

## Enhance Password Validation Logic

### Subtask:
Develop a function, `is_password_strong`, that implements the enhanced password validation rules: minimum length (e.g., 8 characters), presence of at least one uppercase letter, one lowercase letter, one digit, and one special character. This function should return `True` if the password meets all criteria and `False` otherwise.


**Reasoning**:
The subtask requires developing a Python function `is_password_strong` that implements specific password validation rules. This code block defines the function with the required checks for length, uppercase, lowercase, digits, and special characters, and includes example usage to demonstrate its functionality.



````python
import re

def is_password_strong(password):
    """
    Checks if a password meets the enhanced strength criteria.

    Criteria:
    - Minimum length of 8 characters.
    - Contains at least one uppercase letter.
    - Contains at least one lowercase letter.
    - Contains at least one digit.
    - Contains at least one special character (e.g., !@#$%^&*()_+={}[];:'",.<>/?).

    Args:
        password (str): The password string to validate.

    Returns:
        bool: True if the password meets all criteria, False otherwise.
    """
    # 1. Check minimum length
    if len(password) < 8:
        return False

    # 2. Check for at least one uppercase letter
    if not re.search(r"[A-Z]", password):
        return False

    # 3. Check for at least one lowercase letter
    if not re.search(r"[a-z]", password):
        return False

    # 4. Check for at least one digit
    if not re.search(r"[0-9]", password):
        return False

    # 5. Check for at least one special character
    # Using a regex to match any character that is not a letter, digit, or underscore
    if not re.search(r"[^a-zA-Z0-9]", password):
        return False

    # If all checks pass, the password is strong
    return True

# Example usage:
print(f"'Password123!@' is strong: {is_password_strong('Password123!@')}")
print(f"'P@sswOrd1' is strong: {is_password_strong('P@sswOrd1')}")
print(f"'short1!' is strong: {is_password_strong('short1!')}")             # Too short
print(f"'noupper1!' is strong: {is_password_strong('noupper1!')}")         # No uppercase
print(f"'NOLOWER1!' is strong: {is_password_strong('NOLOWER1!')}")         # No lowercase
print(f"'NoDigit!!' is strong: {is_password_strong('NoDigit!!')}")         # No digit
print(f"'NoSpecial123' is strong: {is_password_strong('NoSpecial123')}") # No special character
````
